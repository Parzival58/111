```python
#task-1
class Book():
    def __init__(self,title,author,genre):
        self.__title=title
        self.__author=author
        self.__genre=genre
        self.__availability=True
        self.__borrower=None
    def set_title(self,title):
        self.__title=title
    def get_title(self):
        return self.__title
    def set_author(self,author):
        self.__author=author
    def get_author(self):
        return self.__author
    def set_genre(self,genre):
        self.__genre=genre
    def get_genre(self):
        return self.__genre
    def set_availability(self,available):
        self.__availability=available
    def get_availability(self):
        return self.__availability
    def set_borrower(self,borrower):
        self.__borrower=borrower
    def get_borrower(self):
        return self.__borrower
    def display_info(self):
        if self.__availability==True:
            return (f"{self.__genre} book by {self.__author}, {self.__title} is Available\n"
            "---------------")
        else:
            return (f"{self.__genre} book by {self.__author}, {self.__title} is not Available\n"
            "---------------")
class LibraryMember():
    def __init__(self,id,name):
        self.__member_id=id
        self.__name=name
        self.borrowed_books=[]
    def set_member_id(self,member_id):
        self.__member_id=id
    def get_member_id(self):
        return self.__member_id
    def set_name(self,name):
        self.__name=name
    def get_name(self):
        return self.__name
    def borrow_book(self,book):
        if book.get_availability()==True:
            self.borrowed_books.append(book)
            book.set_availability(False)
            book.set_borrower(self.__member_id)
            print(f"{book.get_title()} has been borrowed by {self.__name}({self.__member_id})")
            print("---------------")
        else:
            print("The requested book is not available")
    def return_book(self,book):
        self.borrowed_books.remove(book)
        book.set_availability(True)
        book.set_borrower(None)
        print(f"{book.get_title()} has been returned by {self.__name}({self.__member_id})")
        print("---------------")
    def display_borrowed_books(self):
        print(f"Books borrowed by {self.__name}:")
        for c in self.borrowed_books:
            print(f"Title: {c.get_title()}\n"
            f"Author {c.get_author()}\n"
            f"Genre:: {c.get_genre()}\n"
            f"Availablity: {c.get_availability()}\n"
            "---------------")
class Library():
    def __init__(self):
        self.books_available=[]
        self.library_members=[]
        print("Created a Library.")
        print("---------------")
    def add_book(self,book):
        self.books_available.append(book)
        print(f"{book.get_title()} by {book.get_author()} has been added to the Library")
        print("---------------")
    def add_library_member(self,member):
        self.library_members.append(member)
        print(f"{member.get_name()} is now a Library Member.")
        print("---------------")
    def display_book_list(self):
        print("All the books in library are:")
        for a in self.books_available:
            print(f"Title: {a.get_title()}\n"
            f"Author {a.get_author()}\n"
            f"Genre:: {a.get_genre()}\n"
            f"Availablity: {a.get_availability()}\n"
            f"Borrowed By: {a.get_borrower()}\n"
            "---------------")
    def display_library_members(self):
        print("The Library Members are:")
        num=1
        for b in self.library_members:
            print(f"{num}.{b.get_name()}")
            num+=1



book1 = Book("Harry Potter and the Chamber of Secrets", "J.K. Rowling", "Fiction")
book2 = Book("Nothing Lasts Forever", "Sidney Sheldon", "Fiction")
book3 = Book("Calculus", "Gilbert Strang", "Education")
book4 = Book("The Great Gatsby", "F. Scott Fitzgerald", "Fiction")
book5 = Book("The Hobbit", "John Ronald Reuel Tolkien", "Fiction")
book6 = Book("One Piece", "Eiichiro Oda", "Manga")
book7 = Book("Dune", "Frank Herbert", "Science fiction")
book8 = Book("Frankenstein", "Mary Shelley", "Science fiction")
book9 = Book("Gone Girl", "Gillian Flynn", "Crime Thriller")
book10 = Book("The Alchemist", "Paulo Coelho", "Novel")
print("======================================")
print(book1.display_info())
print(book2.display_info())
print(book3.display_info())
print(book4.display_info())
print(book5.display_info())
print(book6.display_info())
print(book7.display_info())
print(book8.display_info())
print(book9.display_info())
print(book10.display_info())
member1 = LibraryMember("Member1", "Tom Cruise")
member2 = LibraryMember("Member2", "Brad Pitt")
member3 = LibraryMember("Member3", "Ryan Gosling")
member4 = LibraryMember("Member4", "Christian Bale")
library = Library()
print("======================================")
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)
print("======================================")
library.add_library_member(member1)
library.add_library_member(member2)
print("======================================")
member1.borrow_book(book1)
member1.borrow_book(book2)
member2.borrow_book(book3)
print("======================================")
member3.borrow_book(book3)
print("======================================")
print(book3.display_info())
print("======================================")
library.add_book(book4)
library.add_book(book5)
library.add_book(book6)
print("======================================")
library.add_library_member(member3)
library.add_library_member(member4)
print("======================================")
library.add_book(book7)
library.add_book(book8)
library.add_book(book9)
library.add_book(book10)
print("======================================")
member1.borrow_book(book4)
member2.borrow_book(book5)
member3.borrow_book(book6)
member3.borrow_book(book7)
member4.borrow_book(book8)
member4.borrow_book(book10)
member2.borrow_book(book9)
print("======================================")
library.display_book_list()
print("======================================")
member1.display_borrowed_books()
print("======================================")
member2.display_borrowed_books()
print("======================================")
member3.display_borrowed_books()
print("======================================")
member4.display_borrowed_books()
print("======================================")
member4.return_book(book10)
member3.return_book(book7)
member2.return_book(book5)
member1.return_book(book2)
print("======================================")
member2.borrow_book(book2)
print("======================================")
member1.display_borrowed_books()
print("======================================")
member2.display_borrowed_books()
print("======================================")
member3.display_borrowed_books()
print("======================================")
member4.display_borrowed_books()
print("======================================")
library.display_book_list()
print("======================================")
```
