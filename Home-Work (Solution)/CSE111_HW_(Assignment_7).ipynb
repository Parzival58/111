```python
from datetime import date
class Employee ():
    employee_count={}
    count=0
    def __init__(self,name,joining_date,work_experience,weekly_work_hour):
        self.name=name
        self.joining_date=joining_date
        self.work_experience=work_experience
        self.weekly_work_hour=weekly_work_hour
        if self.weekly_work_hour>60:
            print("Cannot work more than 60 hours per week!")
            self.weekly_work_hour=40
        else:
            self.weekly_work_hour=weekly_work_hour
    @classmethod
    def showDetails(cls):
        # print(cls.employee_count)
        if cls.employee_count=={}:
            print("Company workforce:\n"
                "Total Employee/s: 0")
        else:
            print(f"Total Employee/s : {cls.employee_count['Programmer']+cls.employee_count['HR']}\n"
                f"Total Programmer/s: {cls.employee_count['Programmer']}\n"
                f"Total HR Employee/s: {cls.employee_count['HR']}")

class Programmer(Employee):
    designation_list=[ "Junior Software Engineer","Software Engineer","Senior Software Engineer","Technical Lead"]
    def __init__(self,name,joining_date,work_experience,weekly_work_hour=40):
        super().__init__(name,joining_date,work_experience,weekly_work_hour)
        self.id=self.createProgrammerID()
        self.salary=0
        if "Programmer" in Employee.employee_count.keys():
            Employee.employee_count["Programmer"]+=1
        else:
            Employee.employee_count["Programmer"]=1
        if 0<=self.work_experience<3:
            self.designation="Junior Software Engineer"
        if 3<=self.work_experience<5:
            self.designation="Software Engineer"
        if 5<=self.work_experience<8:
            self.designation="Senior Software Engineer"
        if self.work_experience>=8:
            self.designation="Technical Lead"
    def showProgrammerDetails(self):
        new=self.joining_date[8:]+self.joining_date[4:8]+self.joining_date[:4]
        print(f"Programmer Employee:\n"
            f"Name: {self.name}\n"
            f"ID: {self.id}\n"
            f"Joining Date: {self.joining_date}\n"
            f"Designation: {self.designation}\n"
            f"Salary: BDT {self.salary}\n"
            "-------------")
    def calculateSalary(self):
        if self.designation=="Junior Software Engineer":
            self.salary=30000
        elif self.designation=="Software Engineer":
            self.salary=45000
        elif self.designation=="Senior Software Engineer":
            self.salary=70000
        elif  self.designation=="Technical Lead":
            self.salary=120000
        year=int(self.joining_date[2:4])
        exp=23-year
        for i in range(exp):
            self.salary+=self.salary*15/100
    def createProgrammerID(self):
        count=1
        for i in Employee.employee_count.values():
            count+=i
        self.id="P-"+self.joining_date[2:4]+self.joining_date[5:7]+self.joining_date[8:]+"-"+str(count)
        return self.id
    def calculateOvertime(self):
        if self.weekly_work_hour<=40:
            print(f"{self.name} will not receive any overtime.")
        else:
            amount=(self.weekly_work_hour-40)*2000
            print(f"{self.name} will receive BDT {amount} overtime.")
            self.salary+=amount
        print("-------------")
class HR(Employee):
    def __init__(self,name,joining_date,work_experience,weekly_work_hour=40):
        super().__init__(name,joining_date,work_experience,weekly_work_hour)
        self.id=self.createHREmployeeID()
        if "HR" in Employee.employee_count.keys():
            Employee.employee_count["HR"]+=1
        else:
            Employee.employee_count["HR"]=1
    def showHREmployeeDetails(self):
        new=self.joining_date[8:]+self.joining_date[4:8]+self.joining_date[:4]
        print(f"HR Employee:\n"
            f"Name: {self.name}\n"
            f"ID: {self.id}\n"
            f"Joining Date: {new}")
        print("-------------")
    def createHREmployeeID(self):
        count=1
        for j in Employee.employee_count.values():
            count+=j
        self.id="HR-"+self.joining_date[2:4]+self.joining_date[5:7]+self.joining_date[8:]+"-"+str(count)
        return self.id
class InternProgrammer(Programmer):
    intern_count=0
    def __init__(self,name,joining_date,intern_type="Unpaid"):
        InternProgrammer.intern_count+=1
        self.name=name
        self.joining_date=joining_date
        self.work_experience=0
        self.temp_id="Temp_"+str(InternProgrammer.intern_count)
        self.intern_type=intern_type
        self.work_experience=0
        self.weekly_work_hour=40
        add=8-int(self.joining_date[5:7])
        if add>=4:
            self.status="Eligible for promotion"
        else:
            self.status="Not Eligible for promotion"

    def showInternDetails(self):
        print(f"Intern (Programmer):\n"
            f"Name: {self.name}\n"
            f"ID: {self.temp_id}\n"
            f"Joining Date: {self.joining_date}\n"
            f"Type: {self.intern_type}\n"
            f"Status: {self.status}")
    def promoteToProgrammer(self):
        today = str(date.today())
        if self.status=="Eligible for promotion":
            print(f"{self.name} is promoted!")
            new=Programmer(self.name,today,self.work_experience,self.weekly_work_hour)
            return new
        else:
            print(f"{self.name} can not be promoted.")
Employee.showDetails()
print("==================1==================")
richard = Programmer("Richard Hendricks", "2014-04-19", 6, 50)
dinesh = Programmer("Dinesh", "2016-07-22",4,40)
gilfoyle = Programmer("Bertram Gilfoyle", "2016-07-22", 6, 61)
erlich  = Programmer("Erlich Bachman", "2014-04-19", 9, 5)
bighead = Programmer("Nelson Bighetti", "2017-01-30",3, 41)
print("==================2==================")
richard.calculateSalary()
dinesh.calculateSalary()
gilfoyle.calculateSalary()
erlich.calculateSalary()
bighead.calculateSalary()
print("==================3==================")
richard.showProgrammerDetails()
dinesh.showProgrammerDetails()
gilfoyle.showProgrammerDetails()
erlich.showProgrammerDetails()
bighead.showProgrammerDetails()
print("==================4==================")
richard.calculateOvertime()
dinesh.calculateOvertime()
gilfoyle.calculateOvertime()
erlich.calculateOvertime()
bighead.calculateOvertime()
print("==================5==================")
richard.showProgrammerDetails()
dinesh.showProgrammerDetails()
gilfoyle.showProgrammerDetails()
erlich.showProgrammerDetails()
bighead.showProgrammerDetails()
print("==================6==================")
monica = HR("Monica Hall", "2022-07-06", 2, 40)
russ = HR("Russ Hanneman", "2021-09-14",5,2)
print("==================7==================")
monica.showHREmployeeDetails()
russ.showHREmployeeDetails()
print("==================8==================")
Employee.showDetails()
print("==================9==================")
gavin = Programmer("Gavin Belson", "2016-12-20", 9)
gavin.calculateSalary()
gavin.calculateOvertime()
gavin.showProgrammerDetails()
print("==================10==================")
yang = InternProgrammer("Jian Yang", "2023-01-01")
yang.showInternDetails()
print("==================11==================")
jared = InternProgrammer("Jared Dunn", "2023-06-05", "Paid")
jared.showInternDetails()
print("==================12==================")
jared = jared.promoteToProgrammer()
print("==================13==================")
Employee.showDetails()
print("==================14==================")
yang = yang.promoteToProgrammer()
yang.calculateSalary()
yang.showProgrammerDetails()
print("==================15==================")
Employee.showDetails()
```
