```python
#task-1
class Movie():
    def __init__(self,name,genre,duration):
        self.name=name
        self.genre=genre
        self.duration=duration
    def movieInfo(self):
        return (f"Movie Name:{self.name}\n"
            f"Movie Genre:{self.genre}\n"
            f"Movie Duration:{self.duration} minutes.\n"
            "--------------")
    @classmethod
    def createMovie_fromString(cls,string):
        string=string.split("-")
        name=string[0]
        genre=string[1]
        duration=string[2]
        new=Movie(name,genre,duration)
        return new
class StarCinema():
    all_branch_info={}
    def __init__(self,branch_name):
        self.branch_name=branch_name
        self.movie_list=[]
        StarCinema.all_branch_info[self.branch_name]=self.movie_list
        print(f"Welcome to the {self.branch_name} branch of StarCinema! ")
    def addMovies(self,*movie_objects):
        for i in movie_objects:
            if i not in self.movie_list:
                self.movie_list.append(i)
                print(f"{i.name} added to {self.branch_name} branch. ")
            else:
                print("Movie is already added in this branch. ")
    def removeMovie(self,movieObject):
        self.movie_list.remove(movieObject)
        print(f"{movieObject.name} has been removed from {self.branch_name} branch.")
        print("--------------")
    @classmethod
    def check(cls,string):
        count=0
        for a,b in StarCinema.all_branch_info.items():
            for c in b:
                if string==c.name:
                    print(f"{string} is being streamed in {a} branch.")
                    count+=1
                    genre=c.genre
                    duration=c.duration
        if count!=0:
            print(f"It is of {genre} genre and {duration} minutes duration. ")
        else:

            print(f"{string} is not being streamed in any branch. ")
    @classmethod
    def showAllBranchInfo(cls):
        for j in StarCinema.all_branch_info.keys():
            num=1
            print(f"Branch Name:{j} ")
            for k in StarCinema.all_branch_info[j]:
                print(f"Movie No: {num}\n"
                    f"Movie Name:{k.name}\n"
                    f"Movie Genre:{k.genre}\n"
                    f"Movie Duration:{k.duration} minutes.")
                num+=1
                print("************************** ")
            print("################################# ")
movie1 = Movie('Oppenheimer', 'Biographical Drama', 180)
movie2 = Movie('Avengers: Endgame', 'Action/Sci-fi ', 182)
movie3 = Movie('Thor: Ragnarok', 'Action/Adventure', 130)
movie4 = Movie("Fight Club", "Thriller/Drama", 139)
print('1==========================================')
print(movie1.movieInfo())
print(movie2.movieInfo())
print(movie3.movieInfo())
print(movie4.movieInfo())
print('2==========================================')
movie5 = Movie.createMovie_fromString('The Lord of the Rings: The Return of the King-Fantasy/Adventure-210')
print('3==========================================')
print(movie5.movieInfo())
print('4==========================================')
branch1 = StarCinema('Mohakhali')
print('5==========================================')
branch1.addMovies(movie1, movie2, movie5)
print('6==========================================')
branch1.addMovies(movie3, movie4, movie2)
print('7==========================================')
StarCinema.showAllBranchInfo()
print('8==========================================')
branch2 = StarCinema('Mirpur')
print('9==========================================')
branch2.addMovies(movie2, movie4, movie5)
print('10==========================================')
StarCinema.showAllBranchInfo()
print('11==========================================')
branch3 = StarCinema('Bashundhara City')
print('9==========================================')
branch3.addMovies(movie2, movie3, movie4, movie5)
print('7==========================================')
StarCinema.showAllBranchInfo()
print('8==========================================')
StarCinema.check('Fight Club')
print('12=========================================')
StarCinema.check("Ant man")
print('13=========================================')
StarCinema.check("Thor: Ragnarok")
print('13=========================================')
branch1.removeMovie(movie2)
StarCinema.showAllBranchInfo()
print('14=========================================')
branch2.removeMovie(movie4)
StarCinema.check('Fight Club')
print('14=========================================')
StarCinema.showAllBranchInfo()
print('14=========================================')
branch3.removeMovie(movie2)
branch3.removeMovie(movie5)
StarCinema.showAllBranchInfo()
```
