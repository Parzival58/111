```python
#task-3
class Farmer():
    def __init__(self,name=None):
        self.name=name
        self.land=None
        self.water=None
        self.crops=[]
        self.fishes=[]
        if self.name==None:
            print("Welcome to your farm!")
        if isinstance(self.name, str):
            print(f"Welcome to your farm, {self.name}!")
        if isinstance(self.name, int):
            print(f"Welcome to your farm. Your farm ID is {self.name}! ")
    def addCrops(self,*n):
        self.land=n
        # self.crop=list(self.crop)
        if len(self.land)>0:
            print(f"{len(self.land)} crop(s) added.")
            self.crops.extend(self.land)
        else:
            print("No crop(s) added.")
    def addFishes(self,*m):
        self.water=m
        # self.fish=list(self.fish)
        if len(self.water)>0:
            print(f"{len(self.water)} fish(s) added.")
            self.fishes.extend(self.water)
        else:
            print("No fish(s) added.")
    def showGoods(self):
        if len(self.crops)>0:
            print(f"You have {len(self.crops)} crop(s):")
            index=1
            for i in self.crops:
                if index<len(self.crops):
                    print(i,end=",")
                else:
                    print(i)
                index+=1
        else:
            print("You don't have any crop(s).")
        if len(self.fishes)>0:
            print(f"You have {len(self.fishes)} fish(s):")
            index=1
            for j in self.fishes:
                if index<len(self.fishes):
                    print(j,end=",")
                else:
                    print(j)
                index+=1
        else:
            print("You don't have any fish(s).")



f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute","Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
#task-4
class UniversalStudiosUser():
    def __init__(self,name,age,category):
        self.name=name
        self.age=age
        self.category=category
        self.total=0.0
        print("Welcome to Universal Studios.")
    def selected_rides(self,*n):
        print("Added ride(s) successfully.")
        self.n=n

    def show_details(self):
        print(f"Your information:\n"
            f"Name: {self.name}, Age: {self.age}, Category: {self.category}\n"
            f"Selected rides:")
        for i in self.n:
            i=i.split("-")
            ride=i[0]
            price=int(i[1])
            self.total+=price
            print(f"Ride: {ride}, Amount: {price} dollar(s) ")
        if self.category=="Special" and len(self.n)>3:
            self.total=self.total-self.total*(0.2)
            print("Congratulations!!! You've got a 20% discount.\n"
                f"Please pay {self.total} dollar(s).")
        else:
            print(f"Please pay {self.total} dollar(s). ")



customer_1 = UniversalStudiosUser("Alice", 21, "Special")
print("--------- 1 ---------")
customer_1.selected_rides("Waterworld-100", "Accelerator-200", "DinoSoarin-50")
print("--------- 2 ---------")
customer_1.show_details()
print("=================")
customer_2 = UniversalStudiosUser("Bob", 20, "Normal")
print("--------- 3 ---------")
customer_2.selected_rides("Enchanted Airways-300", "Jurassic Park-500", "Accelerator-200", "DinoSoarin-50")
print("--------- 4 ---------")
customer_2.show_details()
print("=================")
customer_3 = UniversalStudiosUser("Mark", 15, "Special")
print("--------- 5 ---------")
customer_3.selected_rides("Transformers-450", "Jurassic Park-500", "Waterworld-100", "DinoSoarin-50")
print("--------- 6 ---------")
customer_3.show_details()
```

```python
#task-6
class StudentRoutineGenerator():
    def __init__(self,name,num):
        self.name=name
        self.number=num
        self.routine={'Sat/Thurs': {}, 'Sun/Tue': {}, 'Mon/Wed': {}}
        self.new=[]
        self.count=0
        print(f"Name: {self.name}\n"
            f"Maximum Number of Courses: {self.number}")
        print(f"Initial Routine: {self.routine}")
    def addCourses(self,*n):
        for i in n:
            i=i.split("-")
            course=i[0]
            if course not in self.new:
                self.new.append(course)
            else:
                print(f"You already have {course} in your routine")
                continue
            day=i[1]
            time=i[2]
            if day in self.routine:
                if time not in self.routine[day]:
                    if self.count<self.number:
                        self.count+=1
                        self.routine[day][time]=course
                        print(f"Successfully added {course}!")
                    else:
                        print(f"You can't take more than {self.number} courses")
                else:
                    print(f"Can't take {course}. It's clashing with your {self.routine[day][time]} ")
    def showRoutine(self):
        print("Updated routine:\n"
        f"{self.routine}\n"
        "Routine Detains:")
        for key,value in self.routine.items():
            if value!={}:
                print(f"{key}:")
            for time, course in value.items():
                print(f"{time} - {course}")

    def dropCourse(self,course):
        for j in self.routine.values():
            for k,l in j.items():
                if course==l:
                    temp=k
                    flag=True
                    break
                else:
                    flag=False
        if flag==True:
            j.pop(temp)
            print(f"Successfully dropped {course}")
        else:
            print("No such course iin your routine")


print('##################################')
st1 = StudentRoutineGenerator('Harry', 4)
print('------------------------------')
st1.addCourses('CSE110-Mon/Wed-12:30', 'MAT110-Mon/Wed-2:00')
st1.addCourses('ENG101-Sun/Tue-12:30', 'CSE110-Mon/Wed-9:30')
st1.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st1.showRoutine()
print('------------------------------')
st1.dropCourse('CSE110')
st1.dropCourse('PHY112')
print('------------------------------')
st1.showRoutine()
print('##################################')
st2 = StudentRoutineGenerator('John', 3)
print('------------------------------')
st2.addCourses('MAT110-Mon/Wed-8:00')
st2.addCourses('ENG101-Sat/Thurs-12:30', 'CSE110-Sun/Tue-9:30')
st2.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st2.showRoutine()
```
