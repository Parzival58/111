```python
#task-1
class Customer():
    def __init__(self):
        print("Welcome to ABC Memorial Park ")
        self.count=0
        self.price=0
    def buyTicket(self,name,age):

        if self.count<=2:
            self.count+=1
            print(f"Successfully purchased a ticket for {name}!")
            if age>10:
                self.price+=100
            else:
                self.price+=50
        else:
            print(":You can't buy more than 3 tickets ")

    def showDetails(self):
        print(f"Amount of tickets: {self.count}\n"
        f"Total price: {self.price} Taka")


print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
#task-2
class SentenceAnalyzer():
    def __init__(self,sen=None):
        self.sen=sen
        self.count=0
    def set_sentence(self,word):
        self.sen=word
    def get_word_count(self,num=0):
        if num==0:
            self.sentence=self.sen.split(" ")
            self.number=len(self.sentence)
            print(f"Number of words in the sentence: {self.number} ")
        else:
            self.sentence=self.sen.split(" ")
            for i in self.sentence:
                if len(i)==num:
                    self.count+=1
            print(f"Number of words with {num} characters in the sentence: {self.count}")




analyzer1 = SentenceAnalyzer()


analyzer1.set_sentence("That's an easy one")
print("1--------------------------")
analyzer1.get_word_count()
print("2--------------------------")
analyzer2 = SentenceAnalyzer("Like I said it's easy")


print("3--------------------------")
analyzer2.get_word_count()


print("4--------------------------")
analyzer2.get_word_count(4)
print("5--------------------------")
analyzer1.get_word_count(5)
```

```python
#task-3
class Student():
    def __init__(self,name,cgpa,credit=9,dept="CSE"):
        self.name=name
        self.cgpa=cgpa
        self.credit=credit
        self.dept=dept
        self.scholarship=None
    def checkScholarshipEligibility(self):
        if self.credit>10 and self.cgpa>=3.50:
            if self.cgpa>3.70:
                print(f"{self.name} is eligible for Merit-based scholarship.")
                self.scholarship= "Merit-based scholarship"
            if 3.50<=self.cgpa<3.70:
                print(f"{self.name} is eligible for Need-based scholarship.")
                self.scholarship= "Need-based scholarship"
        else:
            print(f"{self.name} is not eligible for scholarship.")
            self.scholarship= "No scholarship"
    def showDetails(self):
        print(f"Name: {self.name}\n"
        f"Department: {self.dept}\m"
        f"CGPA: {self.cgpa}\n"
        f"Number of Credits: {self.credit}\n"
        f"Scholarship Status: {self.scholarship}")



print('--------------------------')
std1 = Student("Alif",3.99,12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim",3.4)
std3 = Student("Henry",3.5,15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob",4,6,"CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
#task-4
class Author():
    def __init__(self,name=None):
        self.name=name
        self.books={}
        self.count=0
    def setName(self,names):
        self.name=names
    def addBook(self,title,genre):
        if self.name==None:
            print("A book can not be added without author name")
        else:
            if genre in self.books:
                self.books[genre].append(title)
                self.count+=1
            else:
                self.books[genre]=[title]
                self.count+=1


    def printDetail(self):
        print(f"Number of Book(s): {self.count}\n"
        f"Author Name: {self.name}")
        for i,j in self.books.items():
            print(i+":",", ".join(j))





a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon","Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```

```python
#task-5
class TaxiLagbe():
    def __init__(self,number,area):
        self.number=number
        self.area=area
        self.count=0
        self.fare=0
        self.person=[]
    def addPassenger(self,*n):

        for i in n:
            info=i.split("_")
            if self.count<=3:
                self.fare+=int(info[1])
                self.count+=1
                self.person.append(info[0])
                print(f"Dear {info[0]}! Welcome to TaxiLagbe.")
            else:
                print("Taxi Full! No more passengers can be added.")
    def printDetails(self):
        print(f"Trip info for Taxi number: {self.number}\n"
            f"This taxi can only cover the {self.area} area.\n"
            f"Total passengers: {self.count}\n"
            "Passenger lists:")
        index=1
        for i in self.person:
            if index==len(self.person):
                print(i)
            else:
                print(i,end=", ")
            index+=1
        print(f"Total collected fare: {self.fare} Taka")




taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200','Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115', 'Parker_215')
print('-------------------------------')
taxi2.printDetails()
```
