```python
#task-01
class Student():
    def __init__(self,name,id,cgpa):
        self.name=name
        self.__id=id
        self.__cgpa=cgpa
    def setId(self,id):
        self.__id=id
    def getId(self):
        return self.__id
    def setCgpa(self):
        self.__cgpa==cgpa
    def getCgpa(self):
        return self.__cgpa
class Department():
    def __init__(self,dept):
        self.dept=dept
        self.info={}
        self.ids=[]
        self.total=[]
    def findStudent(self,id):
        if id not in self.ids:
            print("Student with this ID doesn't exist, Please give a valid ID")
        else:
            print(f"Student info:\n"
                f"Student Name: {self.info[id].name}\n"
                f"ID: {id}\n"
                f"CGPA:  {self.info[id].getCgpa()}")
    def addStudent(self,*n):
        for i in n:
            if i.getId() in self.ids:
                print("Student with the same ID already exists, Please try with another ID")
            else:
                print(f"Welcome to {self.dept} department, {i.name}")
                self.info.update({i.getId():i})
                self.ids.append(i.getId())
                self.total.append(i)
    def details(self):
        print(f"Department Name: {self.dept}\n"
            f"Number of student:{len(self.total)}\n"
            f"Details of the students:")
        for j in self.total:
            print(f"Student name: {j.name}, ID: {j.getId()}, cgpa: {j.getCgpa()}")


s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1,s2,s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib",22301010,3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib",22201010,2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

```python
#task-2
class Spaceship():
    def __init__(self,name,capacity):
        self.name=name
        self.capacity=capacity
        self.car=[]
        self.current=0
        self.total=0
    def load_cargo(self,add):
        self.car.append(add.getMaterial())
        self.current+=add.getWeight()
        if self.current<=self.capacity:
            self.total+=add.getWeight()
        else:
            self.total+=add.getWeight()
            extra=self.total-self.capacity
            print(f"Warning: Unable to load {add.getMaterial()} inside {self.name}. Exceeds capacity by {extra}.")
    def display_details(self):
        print(f"Spaceship Name: {self.name}\n"
        f"Capacity: {self.capacity}\n"
        f"Current Cargo Weight: {self.current}\n"
        f"Cargo: {self.car}")
class Cargo():
    def __init__(self,material,weight):
        self.__material=material
        self.__weight=weight
    def setMaterial(self):
        self.__material=material
    def setWeight(self):
        self.__weight=weight
    def getMaterial(self):
        return self.__material
    def getWeight(self):
        return self.__weight





falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold)  # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium)  # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium)  # This should not exceed Enterprise's capacity
enterprise.display_details()
```

```python
#task-3
class Song():
    def __init__(self,title=None,duration=0,artist="Unknown Artist"):
        self.__title=title
        self.__duration=duration
        self.__artist=artist
        self.min=self.__duration//60
        self.sec=self.__duration%60
    def setTitle(self,title):
        self.__title=title
    def setDuration(self,duration):
        self.__duration=duration
        self.min=self.__duration//60
        self.sec=self.__duration%60
    def setArtist(self,artist):
        self.__artist=artist
    def getTitle(self):
        return self.__title
    def getDuration(self):
        return self.__duration
    def getArtist(self):
        return self.__artist
    def info(self):
        if self.__duration==0 or self.__title==None:
            return ("Cannot create song without title or duration!")
        else:
            return f"{self.__title} by {self.__artist} ({self.min}:{self.sec})"
class Playlist():
    def __init__(self,name,*m):
        self.__name=name
        self.__song_list=list(m)
        self.__now_id=-1
    def setName(self,name):
        self.__name=name
    def setSong_list(self,song_list):
        self.__song_list=song_list
    def setNow_id(self,artist):
        self.__artist=artist
    def getname(self):
        return self.__name
    def getSong_list(self):
        return self.__song_list
    def getNow_id(self):
        return self.now_id
    def play_next(self):
        if self.__now_id==len(self.__song_list)-1:
            print("Reached the end of the playlist, starting from the first song.")
            self.__now_id=0
            print(f"Now playing: {self.__song_list[self.__now_id].info()}")
        else:
            self.__now_id+=1
            print(f"Now playing: {self.__song_list[self.__now_id].info()}")
    def play_prev(self):
        if self.__now_id==0:
            print("Cannot go back from the first song.")
            self.__now_id=0
            print(f"Now playing: {self.__song_list[self.__now_id].info()}")
        else:
            self.__now_id=self.__now_id-1
            print(f"Now playing: {self.__song_list[self.__now_id].info()}")
    def add_songs(self,*songs):
        self.__song_list.extend(songs)
        print(f"Added {len(songs)} song(s) to the playlist {self.__name}")
    def remove_song(self,song_obj):
        print(f"Removed {song_obj.getTitle()} by {song_obj.getArtist()} from the playlist {self.__name}")
        self.__song_list.remove(song_obj)
        self.__now_id=self.__now_id-1
    def show_playlist(self):
        print(f"Playlist: {self.__name}\n"
            f"Total Songs: {len(self.__song_list)}")
        num=1
        for j in self.__song_list:
            print(f"{num}. {j.info()}")
            num+=1
s1=Song("Atlantis", 229, "Seafret")
print(s1.info())
print('===============')
s2=Song("Let Her Go", 252, "Passenger")
print(s2.info())
print('===============')
s3 = Song("Talking To Myself")
print(s3.info())
print('===============')
s3.setTitle("Talking To Myself")
s3.setDuration(231)
s3.setArtist("Linkin Park")
print(s3.info())
print('===============')
s4=Song("Team", 193, "Lorde")
print(s4.info())
print('===============')
s5=Song("Red", 223, "Taylor Swift")
print(s5.info())
print('===============')
p1=Playlist("Liked Song",s1,s2,s3)
p1.show_playlist()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
print('===============')
p1.play_prev()
print('===============')
p1.play_next()
print('===============')
p1.add_songs(s4,s5)
print('===============')
p1.show_playlist()
print('===============')
p1.play_next()
print('===============')
p1.play_next()
print('===============')
p1.remove_song(s1)
print('===============')
p1.show_playlist()
print('===============')
p1.play_next()
print('===============')
```
